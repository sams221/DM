```python
import pandas as pd

fire_alg = pd.read_csv("./DATA/firedataset/viirs-jpss1_2024_Algeria.csv")
fire_tun = pd.read_csv("./DATA/firedataset/viirs-jpss1_2024_Tunisia.csv")

fire_all = pd.concat([fire_alg, fire_tun], ignore_index=True)
print("Shape initial :", fire_all.shape)

fire_all = fire_all[fire_all["confidence"].str.lower() != "low"].copy()
print("Shape après suppression de confidence 'low' :", fire_all.shape)

fire_filtered = fire_all[~fire_all["type"].isin([2, 3])].copy()
print("Shape après suppression de type == 2 et 3 :", fire_filtered.shape)

cols = ["latitude", "longitude"]

fire_filtered = fire_filtered[cols].copy()


print("\nTableau filtré (latitude / longitude) :\n")
print(fire_filtered.head())

output_path = "filtered_fire_data.csv"
fire_filtered.to_csv(output_path, index=False)
print(f"\n✅ Fichier sauvegardé sous : {output_path}")
```
